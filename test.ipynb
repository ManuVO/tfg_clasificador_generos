```python
# Celda de prueba en Jupyter para augmentations y mel spectrograma

import numpy as np
import soundfile as sf
import matplotlib.pyplot as plt
# Librer√≠a de visualizaci√≥n de espectrogramas de librosa (opcional)
import librosa.display  

from features.augment import train_aug
from features.melspectrogram import mel_spec

# 1. Cargar audio de ejemplo (usa un archivo real si lo tienes, o sint√©tico si no)
sr = 22050
y = 0.5 * np.sin(2 * np.pi * 440 * np.linspace(0, 1, sr, endpoint=False)).astype(np.float32)
print("Audio de prueba:", y.shape, "- Duraci√≥n:", len(y)/sr, "segundos")

# 2. Aplicar augmentations y calcular espectrograma Mel
y_aug = train_aug(samples=y, sample_rate=sr)
S = mel_spec(y_aug, sr)
print("Espectrograma Mel (shape):", S.shape)

# 3. Visualizar el espectrograma Mel resultante
plt.figure(figsize=(6, 4))
plt.imshow(S, aspect='auto', origin='lower', cmap='magma')
plt.colorbar(label="Amplitud (dB)")
plt.title("Espectrograma Mel (post-augmentations)")
plt.xlabel("Frames de tiempo")
plt.ylabel("Bines de Mel (frecuencia)")
plt.show()
```
